```python
#Remove Unnecessary columns
Name_Legend = pd.read_csv('Name_Legend.csv')
accepted_relevant_columns = accepted.columns[accepted.columns.isin(Name_Legend['Name'])]
a = accepted[accepted_relevant_columns]
```

```python
#Drop NAs
a = a.dropna(subset=['loan_amnt', 'annual_inc', 'emp_length', 'dti'])
```

```python
#Imput mortgage count NA with 0
a.loc[a['mort_acc'].isnull(), 'mort_acc'] = 0
```

```python
a['issue_d'].head()
```

```python
a['issue_d'] = pd.to_datetime(a['issue_d'], format='%b-%Y')
a['Year'] = a['issue_d'].dt.year
```

```python
a['FICO'] = (a['fico_range_low'] + a['fico_range_high']) / 2
```

```python
a['Loan-Income'] = a['loan_amnt'] / a['annual_inc']
```

```python
a['Return_Total'] = a['total_rec_prncp'] + a['total_rec_int'] + a['total_rec_late_fee']
```

```python
a['ROI'] = a['Return_Total'] / a['loan_amnt'] - 1
```

```python
a['Existing_Debt_Payments_Annual'] = a['dti'] * a['annual_inc']
```

```python
a = a.drop(['issue_d', 'fico_range_low', 'fico_range_high'], axis=1)
```
