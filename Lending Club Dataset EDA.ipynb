```python
missing_accepted = pd.DataFrame({
    'Column': accepted.columns,
    'Missing_Values': accepted.isnull().sum().values.astype(float),
    'Percent_Missing': (accepted.isnull().mean() * 100).values
}).sort_values(by='Missing_Values', ascending=False).reset_index(drop=True)

print('\nMissing Values by Column:\n')
missing_accepted
```

```python
missing_rejected = pd.DataFrame({
    'Column': rejected.columns,
    'Missing_Values': rejected.isnull().sum().values.astype(float),
    'Percent_Missing': (rejected.isnull().mean() * 100).values
}).sort_values(by='Missing_Values', ascending=False).reset_index(drop=True)

print('\nMissing Values by Column:\n')
missing_rejected
```
